```r
library(tidyverse)

download.file("https://github.com/carpentries/assessment/raw/master/learner-assessment/surveys/long_term_survey_current.pdf",
             "survey.pdf")

mydata = read_csv("https://raw.githubusercontent.com/carpentries/assessment/master/learner-assessment/archives/2017/data/longterm_October2017.csv")
#str(mydata)

mydata <- mydata %>%
            select(Country, Position, `Behaviors-Adopted`,
                  Column25:Column33, 
                  `Programming-Usage-Before-Workshop`,
                  `Programming-Usage-Since-Workshop`,
                 `Change-In-Confidence`,
                 `How-Tools-Learned-Help`,
                  Column34:Column38,
                  `Contributed-To-Writing`,
                  `Skills-Advance-Career`,
                  `Improved-Research-Productivity`,
                   `Likely-To-Recommend`
                  )

mydata2 <- mydata %>%
      rename(pre_workshop = `Programming-Usage-Before-Workshop`,
        post_workshop = `Programming-Usage-Since-Workshop`,
        confidence_change = `Change-In-Confidence`,
        writing_skills = `Contributed-To-Writing`,
        career_advancement = `Skills-Advance-Career`,
        research_productivity = `Improved-Research-Productivity`,
        recommend_level = `Likely-To-Recommend` )

country_list <- mydata2 %>%
  count(Country) %>%
  mutate(percent = n * 100 / sum(n)) %>%
  filter(Country != "NA") %>%
  arrange(desc(n))
country_list

prew <- mydata2 %>% 
           count(pre_workshop) %>%
           mutate(pct = n * 100 / sum(n))
           
           

prew <- prew[complete.cases(prew), ]
prew

postw <- mydata2 %>% 
           count(post_workshop) %>%
           mutate(pct = n * 100 / sum(n))
           

postw <- postw[complete.cases(postw), ]
postw

country_list %>%
  ggplot(aes(x = reorder(Country, percent), y = percent)) +
  geom_bar(stat = "identity") +
  coord_flip() +
  labs(x = "Countries of respondents", y = "percentage", title = "Respondents by countries") +
  theme_bw() +
  ggsave("by-countries.png", width = 10, height = 10, units = "in")

combo2 <- combo %>%
            gather("workshop", "values", pre_n, post_n, pre_workshop, post_workshop)

combo2

confidence_change <- mydata2 %>%
                      count(confidence_change) %>%
                      mutate(percent = round(n * 100 / sum(n), 2)) %>%
                      drop_na()
 
library(knitr)
confidence_change %>% kable()

research_productivity <- mydata2 %>%
                      count(research_productivity) %>%
                      mutate(percent = round(n * 100 / sum(n), 2)) %>%
                      drop_na()
 
research_productivity %>% kable()

writing_skills <- mydata2 %>%
                      count(writing_skills) %>%
                      mutate(percent = round(n * 100 / sum(n), 2)) %>%
                      drop_na()
 
writing_skills %>% kable()

workshops <- mydata2 %>%
              select(pre_workshop, post_workshop) %>%
              drop_na() %>%
              mutate(different = (pre_workshop != post_workshop)) %>%
              filter(different == "TRUE") %>%
              filter(pre_workshop == "I had not been using tools like these.")
workshops %>%
    count(post_workshop) %>%
    mutate(pct = n * 100 / sum(n)) %>%
    arrange(desc(pct)) %>%
    ggplot(aes(x = post_workshop, y = pct)) +
    geom_bar(stat = "identity", fill = "blue", colour = "blue") +
    coord_flip() +
    theme_bw() +
    labs(title = "Usage change for people who never used",
        x = "Usage pattern",
        y = "Percent of people who now use") +
    ggsave("usage_change.png", height = 8, width = 10, units = "in")
```
